```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_01.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru


def foo(a):
    """Алгоритм: Располагает элементы списка по возрастанию

    Сложность: O(N + len(i)).
    """
    for i in range(len(a), 0, -1): # O(N)
        for j in range(1, i): # O(len(i))
            if a[j-1] > a[j]: # O(logN)
                a[j-1], a[j] = a[j], a[j-1]
    return a # O(1)


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_02.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru



def foo(i):
    """Алгоритм: Находит количество шагов, необходимых для преобразования числа в строку

    Сложность: O(N)
    """
    digits = "0123456789"
    if i == 0:  # O(1)
        return "0"
    result = ""
    while i > 0: # O(N)
        result = digits[i%10] + result # O(N) - наихудший вариант
        i = i // 10
    return result # O(1)
foo(754)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_03.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru


def foo(s):
    """Алгоритм: Считает сумму всех цифр в указанной строке

    Сложность: O(N^2).
    """
    val = 0
    for c in s: # O(N)
        if c.isdigit(): # O(N) - наихудший вариант
            val += int(c)
    return val # O(1)
foo("Соня54323")
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_04.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru




def foo(n):
    """Алгоритм: Дает список простых чисел до n-ого числа

    Сложность: O(len(n+1)+O(N)).
    """
    res = []
    for i in range(1, n + 1): # O(len(n+1))
        divisors = 0
        j = 2
        while j < i and divisors == 0: # O(N)
            if i % j == 0: # O(log(N))
                divisors += 1
            j += 1

        if divisors == 0: # O(log(N))
            res.append(i) # O(log(N)) - наихудший вариант

    return res # O(1)
foo(30)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_05.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru


def foo(nums):
    """Алгоритм: Есть ли в приведенном списке четное число

    Сложность: O(N^2).
    """
    for x in nums: # O(N)
        if x % 2 == 0: # O(N)
            return True # O(1) - наихудший вариант
    else:
        return False # O(1)
foo([45,35,43,24,24])
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_06.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru


def foo(nums):
    """Алгоритм: Первый элемент в списке суммируется с последним элементом списка в квадрате

    Сложность: O(1).
    """
    return (nums[0] + nums[-1] ** 2) # O(1)
foo([34,23,34,25])
```
