```python
# 2

class Person: # создание класса Персона
  def __init__(self,surname,name,patronymic,country, experiences): # конструктор
    self.name = name
    self.surname = surname
    self.patronymic = patronymic
    self.country = country
    self.experiences = experiences
  def __str__(self):
    return "Фамилия: {}, имя: {}, отчество: {}, страна:, {}\nОпыт работы: {}\nРаботник является {}\nДолжность: {}\nзарплата: {}\nКоличество рабочих дней в неделю: {}\nРаботник имеет стаж работы {}\n".format(self.surname, self.name,self.patronymic,self.country,self.experiences,self.countrys(),self.post(),self.salary(),self.work(), self.experience())

  def post(self): # функция-метод 1: должность
   return "Не указана"
  def salary(self): # функция-метод 2: зарплата
    return "Не имеет"
  def work(self): # функция-метод 3: количество рабочих дней в неделю
    return "0"
  def countrys(self):
    if self.country == "Россия":
      return "местным"
    else:
      return "иностранным"
  def experience(self):
    if 0<= self.experiences < 3:
      return "маленький"
    elif 3 <= self.experiences <= 9:
      return "средний"
    elif self.experiences > 9:
      return "большой"
    else:
      return "неизвестно"

class Employee(Person): # создание класса Служащий, производный от Персоны
  def __init__(self,name,surname,patronymic,country, experiences): 
    super().__init__(name,surname,patronymic,country, experiences) # наследование параметров от класса Person
  def post(self): # функция-метод 1: должность
    return "Служащий"
  def salary(self): # функция-метод 2: зарплата
    return "150-300 тысяч рублей"
  def work(self): # функция-метод 3: количество рабочих дней в неделю
    return "6"

class Worker(Employee): # создание класса Рабочий, производный от Служащего
  def __init__(self,name,surname,patronymic,country, experiences):
    super().__init__(name,surname,patronymic,country, experiences) # наследование параметров от класса Person
  def post(self): # функция-метод 1: должность
    return "Рабочий"
  def salary(self): # функция-метод 2: зарплата
    return "30-50 тысяч рублей"
  def work(self): # функция-метод 3: количество рабочих дней в неделю
    return "4"

class Engineer(Employee): # создание класса Инженер, производный от Служащего
  def __init__(self,name,surname,patronymic,country, experiences):
    super().__init__(name,surname,patronymic,country,experiences) # наследование параметров от класса Person
  def post(self): # функция-метод 1: должность
    return "Инженер"
  def salary(self): # функция-метод 2: зарплата
    return "60-80 тысяч рублей"
  def work(self): # функция-метод 3: количество рабочих дней в неделю
    return "5"


a1 = Person("Морозова", "Анна", "Валерьевна","Россия", 2)
print(a1)
a2 = Employee("Потапов", "Михаил", "Дмитревич", "Беларусь", 10)
print(a2)
a3 = Worker("Смирнова", "Марина", "Владимировна", "Россия", 5)
print(a3)
a4 = Engineer("Иванов", "Александр", "Андреевич", "Украина", -6)
print(a4)
```
