```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_07_02_02.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru


"""
Ошибки (номера строк через пробел:, данная строка - №2): 10 11 12 14
"""


def primes(a, b):
    """Вернуть список простых чисел на отрезке от 'a' до 'b'."""
    res = [] # создание списка

    for i in range(a, b+1):
      c=2
      while i % c != 0: # если остаток от деления равен 0
        c += 1 # перейти к проверке следующего числа
      if c == i: # если остаток от деления не равен 0
        res.append(i) # добавить в список число

    return res

primes(2,7)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_07_02_04.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru



"""
Ошибки (номера строк через пробел, данная строка - №2): 8 9 10 11
"""


def min_pair(nums):
    """Вернуть минимальную сумму соседних 2-х чисел в списке 'nums'."""
    min = float("inf") # создание вещественной переменной
    for i in range(len(nums)-1): # выполнить цикл для заданного количества чисел в списке
        if nums[i] + nums[i + 1] <min: # если сумма двух соседних чисел меньше предыдущей суммы соседних чисел
          min=nums[i] + nums[i + 1] # записать сумму чисел

    return min

import random

random.seed(10) # генератор рандомных чисел

N_MAX = 10
RANGE_MIN = 1
RANGE_MAX = 100
nums = random.sample(range(RANGE_MIN, RANGE_MAX), N_MAX)

print(nums)

print(min_pair(nums))
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_07_02_07.
#
# Выполнил: Шведова Софья Сергеевна
# Группа: АДЭУ-211
# E-mail: shvedova_sofiya23@mail.ru




"""
Ошибки (номера строк через пробел, данная строка - №2): 11 17 18 
"""


def power(x, y=2):
    """Вернуть x^y."""
    if y == 0: # если степень равна 0
        return 1 # вернуть 1
    elif type(y) == int: # если y - целое число
        return x * power(x, y - 1)
    else: # если не выходит посчитать степень
        raise ValueError("Невозможно посчитать возведение в степень "
                         "введенных чисел.")

try: # попытаться выполнить вычисления
  x = int(input("x="))
  y = int(input("y="))
  print(power(x, y))
except ValueError as err: # если не получилось выполнить вычисления
    print("Ошибка:", err, ". Проверьте введенные числа.")
except Exception as err:
    print("Ошибка:", err)
```
